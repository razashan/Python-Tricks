```python
mba = pd.read_excel('Items.xlsx')
```

```python
mba.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = list(mba["Items"].apply(lambda x:x.split(",") ))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df = pd.DataFrame(a_data,columns=a.columns_)
df = df.replace(False,0)
df
```

```python
df = apriori(df, min_support = 0.2, use_colnames = True, verbose = 1)
df
```

```python
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.1)
df_ar
```

#### From the above results of Association Rule Mining we can say that if the Broccoli is picked the chances of Carrot being picked ind 36% and If the Carrot is picked the chances of Broccoli is being picked is 51%
